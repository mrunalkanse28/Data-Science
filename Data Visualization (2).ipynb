```python
x = pd.Series(range(1,20,2))
y = np.log(x)
plt.figure(figsize=(7,8)) # 7 define length and 8 define height
plt.plot(x,y,color = 'green',marker='*',linestyle='--') # marker used for highlight coordinates and linestyle used for type of line
plt.title('Log of first 10 even numbers',weight='bold')
plt.xlabel('Numbers')
plt.ylabel('Log of numbers')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.legend('Log Function')
plt.show()
```

```python
data = {
 'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
 'Stock Price': [150, 155, 160, 162, 158, 165, 170, 172, 168, 175, 180, 185]
}
df = pd.DataFrame(data)
plt.figure(figsize=(6,3))
plt.plot(df['Month'],df['Stock Price'],marker='o',color='b')
sns.set_style('darkgrid')
plt.show()
```

```python
data = {
 'Year': [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012],
 'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
 'Stock Price': [150, 155, 160, 162, 158, 165, 170, 172, 168, 175, 180, 185]
}
df = pd.DataFrame(data)

plt.figure(figsize=(6,3))
sns.scatterplot(x = 'Month',y = 'Stock Price',data=df,hue='Year',palette='Set2',markers='o')
plt.legend(title='Year',bbox_to_anchor=(1,1),loc='upper left')
plt.show()
```
