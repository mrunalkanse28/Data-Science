#### WEB SCRAPING

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Which director has highest no of movies in top 50 high rated movies
# Which joner has the highest rating among the top 50 rated movies
# 
```

```python
url = 'https://www.imdb.com/search/title/?title_type=feature&release_date=1950-01-01,2012-12-31&sort=num_votes,desc'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'} 

response = requests.get(url,headers= headers)
if response.status_code == 200:
    page_html = response.text
    print(f'page fetched sucessfully')
else:
    print(f'failed to fetch page')
    exit()
```

```python
page_html[:1000]
```

```python
page_soup = BeautifulSoup(page_html, 'html.parser')
page_soup
```

```python
containers = page_soup.find_all('li',{'class':'ipc-metadata-list-summary-item'})
print(f'Number of containers found: {len(containers)}')
```

```python
if len(containers) > 0:
    filename = 'imdb_movies.csv'
    f = open(filename, 'w')


    headers_file = 'Name,Year,Duration\n'
    f.write(headers_file)
    for container in containers:
        name_mov = container.find_all('h3',{'class':'ipc-title__text'})
        Name = name_mov[0].text.strip() if name_mov else 'N/A'

        year_mov = container.find_all('span',{'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})
        Year =  year_mov[0].text.strip() if year_mov else 'N/A'

        duration_mov = container.find_all('span',{'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})
        Duration =  duration_mov[1].text.strip() if duration_mov else 'N/A'

        print(f'{Name},{Year},{Duration}')
        f.write(f'{Name},{Year},{Duration}\n')

    f.close()
```
