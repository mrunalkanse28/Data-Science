```python
# You are given a data dictionary with students_info of 5 students. Solve the questions that follow.
data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'MathScore': [85, 92, 78, 90, 88],
    'EnglishScore': [78, 81, 86, 77, 93],
    'ScienceScore': [89, 95, 80, 85, 91]
}
```

```python
# Create a DataFrame using the given data dictionary.
data1 = pd.DataFrame(data)
data1
```

```python
# Calculate the average score for each student and add it as a new column to the DataFrame.
data1['AverageScore'] = data1[['MathScore','EnglishScore','ScienceScore']].mean(axis = 1)
data1
```

```python
# Find the student with the highest average score.
highest_average = data1['AverageScore'].idxmax()
stud_with_highest_score = data1.loc[[highest_average]] # here if we want data in table format then add double square bracket
stud_with_highest_score
```

```python
# Filter and display students who scored more than 85 in Science.
science_topper_stud = data1[data1['ScienceScore'] > 85]
science_topper_stud
```

```python
# Calculate the mean Math score for the class.
average_math = data1['MathScore'].mean()
print('The Total Average of maths Score is',average_math)
```

```python
# Create a new column that categorizes students based on their average score (e.g., 'High' for >85, 'Medium' for 70-85, and 'Low' for <70).
data1['Category'] = data1['AverageScore'].apply(lambda x: 'High' if x > 85 else ('Medium' if x >= 70 else 'Low'))
data1
```

```python
# Sort the DataFrame by English scores in ascending order.
sorted_df = data1.sort_values(by='EnglishScore')
sorted_df
```

```python
# Calculate the correlation matrix for the Math, English, and Science scores.
correlation_matrix = data1[['MathScore','EnglishScore','ScienceScore']].corr()
correlation_matrix
```

```python
# Add a new student to the DataFrame.
data1 = pd.DataFrame(data)
new_student = {
    'StudentID': 6,
    'Name': 'Frank',
    'MathScore': 67,
    'EnglishScore': 55,
    'ScienceScore': 55
}

data1.loc[len(data1)] = new_student
print(data1)

data1['AverageScore'] = data1[['MathScore','EnglishScore','ScienceScore']].mean(axis = 1)
data1

data1['Category'] = data1['AverageScore'].apply(lambda x: 'High' if x > 85 else ('Medium' if x >= 70 else 'Low'))
data1
sorted_df = data1.sort_values(by='EnglishScore')
sorted_df
```

```python
# Remove the student with the lowest average score from the DataFrame.
index_of_low_average = data1['AverageScore'].idxmin()
index_of_low_average
```

```python
data1 = data1.drop(index_of_low_average).reset_index(drop=True)
print(data1)
```
