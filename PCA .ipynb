```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('digits.csv')
digits.head()
```

```python
new_df = digits.drop('number_label',axis=1)
new_df
```

```python
first_row = new_df.iloc[0]
```

```python
first_row
```

```python
first_row_array = first_row.to_numpy()
first_row_array
```

```python
reshape_array = first_row_array.reshape(8,8)
reshape_array
```

```python
sns.heatmap(reshape_array)
plt.show()
```

```python
plt.imshow(reshape_array, cmap='viridis')
plt.show()
```

```python
sns.heatmap(reshape_array,cmap='gray',annot=True)
plt.show()
```
