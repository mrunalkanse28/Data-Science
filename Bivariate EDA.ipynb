### Bivariate EDA

```python
# It involves two variables. Focusses on understanding the relationship between two variables using co-relation
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bank_data_clean.csv')
df.head()
```

```python
df.columns
```

#### Numeric_numeric EDA:
1. Scatterplot
2. Pairplot
3. Correlation

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df.age, y=df.salary)
```

```python
correl = df[['age','salary']].corr()
correl
```

```python
correl1 = df[['balance','salary']].corr()
correl1
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df.balance, y=df.salary)
```

```python
correl1 = df[['age','balance','salary']].corr()
correl1
```

```python
plt.figure(figsize=(20,10))
sns.pairplot(data=df,vars=['age','balance','salary'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correl1, annot=True, cmap='Reds') # annot=True is used for printing values of correlation and cmap is used for colour
plt.show()
```

### Numeric-Categorical data EDA:
1. Boxplot
2. Barplot

```python
df1 = df.groupby('response')['salary'].mean()
df1
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df,x='response',y='salary')
plt.show()
```

```python
df1 = df.groupby('response')['salary'].median()
df1
```

```python
df1 = df.groupby('response')['salary'].quantile(0.25)
df1
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df,x='response',y='balance')
plt.show()
```

```python
df1 = df.groupby('response')['balance'].median()
df1
```
