## Line plot:
* A line plot is a graphical representation that illustrates the trend of a variable over a specified period.
* It is particularly useful for tracking changes and trends over time.
* This plot type is versatile, capable of handling both numerical and categorical data effectively.

1. Purpose:

 Trend Visualization: Line charts are ideal for showing how a variable evolves over time, making it easy to identify patterns, trends, and fluctuations.


2. Components:

 * X-Axis (Horizontal): Typically represents time intervals (e.g., days, months, years) or categories if dealing with categorical data.
 * Y-Axis (Vertical): Represents the variable being measured (e.g., sales figures, temperature, stock prices).


3. Data Points and Lines:

 * Data Points: Individual values plotted on the chart, corresponding to specific time intervals or categories.
 * Lines: Connecting these points to show the continuous trend of the variable.


4. Types of Data:

 * Numerical Data: Direct measurements or calculations, such as revenue, temperature, or stock prices.

 * Categorical Data: Grouped data that can be organized into categories, such as sales by product type or performance by department.


5. Example Use Case
 * Numerical Data: Tracking monthly sales revenue over a year. The X-axis would represent months (January to December), and the Y-axis would represent revenue amounts. The line connecting the data points shows whether sales are increasing, decreasing, or remaining stable over time.

 * Categorical Data: Monitoring the performance of different product categories over quarters. The X-axis could represent quarters (Q1, Q2, Q3, Q4), while the Y-axis represents performance metrics (e.g., number of units sold). The line shows how each category's performance trends over the quarters.

### Benefits
1. Clarity: Offers a clear visual representation of trends, making it easy to interpret complex data.
2. Comparison: Multiple lines can be plotted to compare different variables or categories simultaneously.
3. Insights: Helps in identifying patterns such as seasonality, growth, or decline, aiding in decision-making processes.

## Bar plot:
* A bar plot is a visual representation of data where individual bars represent different categories or groups.
* The length or height of each bar is proportional to the value it represents, making it easy to compare different categories at a glance.
* Bar charts are widely used in various fields for their simplicity and effectiveness in conveying information.

1. Purpose

 Comparison: Bar graphs are primarily used to compare values across different categories or groups.


2. Axes:

 * X-Axis (Horizontal): Represents the categories or groups being compared. These can be numerical ranges, categorical data, or time intervals.

 * Y-Axis (Vertical): Represents the values associated with each category or group. These values can be frequencies, counts, percentages, or any other measurable quantity.


3. Bars:

 * Height/Length: Proportional to the value it represents on the Y-axis.

 * Width: Usually uniform across all bars for consistency.

 * Orientation: Bars can be vertical (most common) or horizontal, depending on the data and the desired presentation.


4. Data Points

 * Categories: Represented on the X-axis, these are the groups or types being compared.

 * Values: Represented on the Y-axis, these are the quantities measured for each category.

 * Individual Bars: Each bar represents a single data point, combining a category and its corresponding value.

5. Types of Data

 * Categorical Data: Non-numerical data that can be divided into distinct groups, such as product types, regions, or survey responses.

 * Numerical Data: Data that represents counts or measurements, such as sales figures, population numbers, or test scores.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example 2
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']*2,
    'Rainfall': [78.1, 92.3, 101.5, 76.2, 89.4, 120.5, 140.2, 135.6, 110.4, 95.3, 85.1, 79.2, 88.1, 62.3, 90.5, 56.2, 99.4, 130.5, 120.2, 115.6, 118.4, 105.3, 75.1, 89.2],
    'Region': ['Region 1']*12 + ['Region 2']*12
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(6,3))  # Adjust the figure size for better readability
sns.barplot(x='Month', y='Rainfall', data=df, palette="viridis", hue='Region')


# Add a title
plt.title("Monthly Rainfall")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1,1), loc='upper left')

# Show the plot
plt.show()
````

In the above example we used "hue", why is it useful?

The hue parameter in Seaborn's barplot function is used to add a third dimension to the plot, allowing for an additional categorical variable to be visualized. This parameter **groups** the data by different colors within each category on the x-axis, making it easier to distinguish between subcategories or to compare multiple datasets within each main category.


Let us see how exactly hue is useful:

````python
# Example 3:
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] * 2,
    'Year': ['2023'] * 12 + ['2024'] * 12,
    'Rainfall': [78.1, 92.3, 101.5, 76.2, 89.4, 120.5, 140.2, 135.6, 110.4, 95.3, 85.1, 79.2,
                 70.2, 85.6, 93.4, 81.8, 88.9, 110.2, 130.5, 125.8, 102.3, 90.5, 80.9, 75.1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bar plot with hue='Year'
plt.figure(figsize=(6, 3))
sns.barplot(x='Month', y='Rainfall', hue='Year', data=df)
plt.legend(title='Year', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
````

## Scatter plot:
A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It consists of individual data points, each representing an observation in the dataset. Scatter plots are particularly useful for identifying patterns, correlations, and trends between variables.


1. Purpose
 * Relationship Exploration: Scatter plots are used to visualize the relationship between two numerical variables and identify any patterns or trends.
 * Correlation Analysis: They help in assessing the strength and direction of the relationship between variables.
 * Outlier Detection: Scatter plots can also be used to identify outliers or anomalies in the data.

2. Components
 * X-Axis (Horizontal):
 Represents one of the numerical variables being compared.
 * Y-Axis (Vertical):
Represents the other numerical variable being compared.

3. Data Points
 * Coordinates: Each data point is represented by a pair of coordinates (x, y), where x is the value of the variable on the x-axis and y is the value of the variable on the y-axis.

4. Types of Data
 * Numerical Data: Scatter plots are most commonly used to visualize the relationship between two continuous numerical variables. For example, plotting height against weight, or temperature against sales volume.

[Scatter plot using matplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter)



```

```



````python
# Example 1
x =[5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y =[99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

plt.figure(figsize=(6,3))
plt.scatter(x, y)
plt.show()
````

Customization of scatter plot: alpha, c

The alpha parameter controls the transparency of the points in the scatter plot. It takes a float value between 0 and 1, where:

* 0 makes the points completely transparent (invisible).
* 1 makes the points completely opaque (fully visible).
* Any value in between adjusts the transparency accordingly. For example, 0.5
would make the points 50% transparent.

The c parameter controls the color of the points in the scatter plot. It can take various types of inputs:

* Single color format string: All points will be of the same color.
* Sequence of color specifications: Each point can have a different color.
* Array of numbers: This array will be mapped to colors using a colormap.
Using c allows for customization of point colors to represent different categories or to map to another variable's value.

````python
x =[5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y =[99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

plt.figure(figsize=(6,3))
plt.scatter(x, y, c ="red",alpha=0.5)
plt.show()
````

````python
# Example 3: Creation of dataframe which is relevant to show the scatter plot uses

# Define cluster centers
cluster_centers = np.array([[70, 80, 75, 5],    # Cluster 1 (High performers, study more)
                             [50, 60, 55, 3],   # Cluster 2 (Medium performers, study moderate)
                             [30, 40, 35, 2]])  # Cluster 3 (Low performers, study less)

# Define cluster standard deviations
cluster_std = np.array([[5, 7, 8, 2],        # Cluster 1
                         [8, 10, 12, 2],     # Cluster 2
                         [10, 12, 15, 2]])   # Cluster 3

# Number of data points
num_points = 300

# Generate data points within clusters
data_points = np.empty((0, 4))

for i in range(len(cluster_centers)):
    data_points = np.vstack((data_points,
                             np.random.normal(loc=cluster_centers[i],
                                              scale=cluster_std[i],
                                              size=(num_points//3, 4))))

# Introduce outliers
num_outliers = num_points // 20
outliers = np.random.uniform(low=[0, 0, 0, 0], high=[100, 100, 100, 10], size=(num_outliers, 4))
data_points = np.vstack((data_points, outliers))


# Create DataFrame
data = pd.DataFrame(data_points, columns=['Math', 'Science', 'Literature', 'Study Hours'])
data
````

````python
# Plot scatter plot
sns.scatterplot(x='Math', y='Science', hue='Study Hours', size='Study Hours', data=data, palette='viridis', sizes=(5, 100))
plt.show()
````

From the above chart, you can easily see that students who studied for more hours scored higher in both subjects. In the absence of the chart, it becomes more difficult to glean insights quickly.

## Histogram:
1. Purpose:

 A histogram is used to visualize the distribution of a dataset. It shows the frequency of values within predefined intervals, known as bins, along the range of the data.

2. Components:

 * Bins: Intervals along the range of the data where values are grouped.
 * Frequency/Count: Number of data points falling within each bin.
 * X-axis: Represents the range of the data.
 * Y-axis: Represents the frequency or count of data points within each bin.

3. Bars:

 Rectangular bars drawn for each bin, whose height represents the frequency/count of data points in that bin.

4. Data Points:

 Histograms work with numerical data. Each data point represents a value from the dataset.

5. Types of Data:

 Histograms are suitable for analyzing the distribution of **continuous** data. They are commonly used in fields such as statistics, finance, and science to explore the shape, center, and spread of a dataset.

Example Use Case:
Suppose you have a dataset of exam scores for a class. You can create a histogram to visualize the distribution of scores and observe patterns such as whether the scores are normally distributed, skewed, or have outliers.


**Benefits:**

* Visual Representation: Histograms provide a visual representation of the distribution of data, making it easier to understand than raw data.
* Identify Patterns: They help identify patterns, such as central tendency, variability, and presence of outliers, in the dataset.
* Data Exploration: Histograms allow for quick data exploration and hypothesis generation, guiding further analysis.

[Hist plot using matplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib-pyplot-hist)



```
plt.hist(x, bins=None, range=None, etc.,)
```



````python
# Generate synthetic exam scores data
exam_scores = np.random.normal(loc=70, scale=15, size=100)  # Normally distributed scores

# Create a histogram using Matplotlib
plt.figure(figsize=(6, 3))
plt.hist(exam_scores, bins=25, color='skyblue', edgecolor='skyblue')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic exam scores data (clipped to not exceed 100)
exam_scores = np.clip(np.random.normal(loc=70, scale=15, size=100), a_min=None, a_max=100)

# Create a histogram using Matplotlib
plt.figure(figsize=(6, 3))
plt.hist(exam_scores, bins=10, range=(50,100), color='skyblue', edgecolor='skyblue')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')

plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)
plt.show()
````

### Boxplot:
Box plots divide the data range into three important categories, which are as follows:

* Median value: This is the value that divides the data range into two equal halves, i.e., the 50th percentile.

* Interquartile range (IQR): These data points range between the 25th and 75th percentile values.

* Outliers: These are data points that differ significantly from other observations and lie beyond the whiskers.

image.png

1. Purpose:

 A boxplot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is used to identify the central tendency, variability, and skewness of a dataset, as well as to detect outliers.


2. Components:
* Box: The central box represents the interquartile range (IQR), which contains the middle 50% of the data.
* Whiskers: The lines (or "whiskers") extending from the box indicate the range of the data, typically up to 1.5 times the IQR from the quartiles.
* Median Line: A line inside the box indicates the median (Q2) of the data.
* Outliers: Individual points outside the whiskers represent outliers, which are unusually high or low values.
* Minimum and Maximum: The smallest and largest data points within 1.5 times the IQR from the quartiles.
* Data Points:
Each data point in a boxplot represents a value from the dataset. The dataset is divided into quartiles to construct the box and whiskers.


3. Types of Data:
Boxplots are suitable for continuous numerical data. They can be used for single-variable analysis or for comparing distributions across different categories.


4. Example Use Case:
Suppose you have a dataset of salaries for employees in different departments. You can create boxplots to compare the salary distributions across departments, identifying the central tendency, spread, and any potential outliers in each department.


Benefits:

* Summary Visualization: Boxplots provide a concise summary of a dataset's distribution, highlighting the central tendency, spread, and skewness.
Outlier Detection: They effectively identify outliers, helping to pinpoint unusual data points that may require further investigation.
* Comparison Across Groups: Boxplots facilitate the comparison of distributions across multiple groups or categories.
* Non-parametric: Boxplots do not assume any specific data distribution, making them useful for a wide range of datasets.
* Insight into Data Symmetry: By showing the quartiles and median, boxplots help assess the symmetry or skewness of the data distribution.

### Piecharts:

1. Purpose
 A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions. It is used to show the relative sizes of parts to a whole, representing categories as slices of a pie.

2. Components
 * Circle: The overall shape representing the whole dataset.
 * Slices: Segments of the circle, each representing a category's proportion relative to the whole.
 * Labels: Text annotations that identify each slice, often showing the category name and/or the percentage value.
 * Legend: A key that explains the colors or patterns used for each slice (optional).

3. Data Points
 * Each slice corresponds to a single data point that represents a category.
 * The size of each slice is proportional to the value of the corresponding data point.
 * Values are often expressed as percentages of the total sum of all data points.

4. Types of Data
 * Categorical Data: Pie charts are best suited for categorical data where each category represents a part of the whole.
 * Nominal Data: Data without a specific order (e.g., types of fruits, regions).
 * Ordinal Data: Data with a specific order, although less common for pie charts.

Example Use Case

Market Share Analysis:

Suppose you want to analyze the market share of different companies in the smartphone industry. You can create a pie chart to show how much market share each company holds relative to the total market

[pie plot using matplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib-pyplot-subplots)



```
plt.pie(x, explode=None, labels=None, colors=None, autopct=None, etc.,)

```



````python
# Example 1
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [40, 30, 20, 10]

# Plot
plt.figure(figsize=(4,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
````

Now, let us add customization to it:

````python
# Example 2
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [40, 30, 20, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice (Company A)


plt.figure(figsize=(4,4))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Market Share of Smartphone Companies')
plt.show()
````
