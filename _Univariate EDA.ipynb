Univariate is derived from 'Uni' meaning single and 'Variate' refers to variable.

Data is of two types: Numerical and Categorical. Univariate EDA for both the types are different. For Categorical we can use count, percentage of each category and for visualization we can use bar chart, pie charts , etc. For numerical Data univariate EDA includes mean,median,mode,std,dev,variance,quartiles,min,max(Descriptive statistics) and visualization includes histogram,boxplot,pie charts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bank_data_clean.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('pdays',axis=1,inplace=True)
df.isnull().sum()
```

```python
df.drop('Unnamed : 0',axis=1,inplace=True)
```

### Histogram of age column

```python
plt.figure(figsize=(10,8))

plt.hist(df.age, bins = 20)
plt.title('AGE DISTRIBUTION')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Boxplot of age column

```python
plt.figure(figsize=(6,4))
sns.boxplot(x =df.age)
plt.title('AGE DISTRIBUTION')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Pie chart of marital column

```python
label=['Married','Single','Divorced']
data=[60,28,12]
plt.figure(figsize=(4,4))
plt.pie(data,labels = label,autopct='%1.1f%%')
plt.show()
```
